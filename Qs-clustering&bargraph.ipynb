```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
Qs = data2[['ar score', 'er score', 'fsr score',  'cpf score', 'ifr score', 'isr score',
                'irn score', 'ger score', 'score scaled']]
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(Qs) #데이터 학습
scaler_data = scaler.transform(Qs) #변환
```

```python
pca = PCA(n_components = 2) #2차원 시각화를 위해 변수 2개 설정
```

```python
pca.fit(scaler_data)
```

```python
data2 = pd.DataFrame(data = pca.transform(scaler_data), columns=['pc1', 'pc2']) #PCA로 scaler_data를 학습시킨다
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# create model and prediction
for i in range(2, 7):
  model = KMeans(n_clusters=i,algorithm='auto')   #cluster개수 변경하고 싶으면
  model.fit(data3)
  predict = pd.DataFrame(model.predict(data3))
  predict.columns=['labels']
  r = pd.concat([data3,predict],axis=1)
  r2 = r[r.index < 50] 
  r2
```

```python
narray_value  #선수별 군집 어디에 해당되는지     
ppc1_value
ppc2_value
centers = pd.DataFrame(model.cluster_centers_,columns=['pc1','pc2'])
center_x = centers['pc1']
center_y = centers['pc2']
   
# scatter plot  # 중심점
plt.scatter(r2['pc1'],r2['pc2'],c=r2['labels'],alpha=0.5)
plt.scatter(center_x,center_y,s=50,marker='D',c='r')   
plt.xlabel("pc1")
plt.ylabel("pc2") 
for i, txt in enumerate(narray_value):
    plt.annotate(txt, (ppc1_value[i], ppc2_value[i]), fontsize = 8)
      
plt.show()
```

```python
# n 값을 늘려가며 반복 테스트
for i in range(3, 7):
        estimator = KMeans(n_clusters = i)
        ids = estimator.fit_predict(data3[['pc1', 'pc2']])
        centers = pd.DataFrame(estimator.cluster_centers_,columns=['pc1', 'pc2'])
        center_x = centers['pc1']
        center_y = centers['pc2']
        plt.title("Cluster n = {}".format(i))
        plt.xlabel('pc1')
        plt.ylabel('pc2')
        for i, txt in enumerate(narray_value):
            plt.annotate(txt, (ppc1_value[i], ppc2_value[i]), fontsize = 8, alpha = 0.3)
        plt.scatter(data3['pc1'], data3['pc2'], c=ids)  
        plt.figure(figsize = (6,5))
plt.show()
```
